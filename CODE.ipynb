```python
import pandas as pd
import cupy as cp
import cupyx.scipy.sparse as cs
from cupyx.scipy.sparse import csr_matrix, hstack, vstack
import numpy as np
import pandas as pd
import numpy as np
import time
def get_OD_vector(test_folder, sample_od, txt_files):
    """
    Create an Origin-Destination (OD) vector from a list of text files.

    Parameters:
        test_folder (str): The folder containing the text files.
        sample_od (DataFrame): Sample OD matrix for initialization.
        txt_files (list): List of text files to process.

    Returns:
        DataFrame: An OD vector DataFrame with columns ['origin', 'destination', 'flow', 'ent'].
    """
    df_list = []

    # Initialize the sample_od DataFrame with zeros
    sample_od = sample_od.fillna(0) * 0

    for i, txt_file in enumerate(txt_files):
        # Read the data from the current text file
        df1 = pd.read_csv(txt_file, sep='\t', header=0, index_col=0)
        
        # Ensure column names are of type 'int64'
        df1.columns = df1.columns.astype('int64')
        
        # Add sample_od to the current data, fill NaN with zeros, and reshape the DataFrame
        df1 = (df1 + sample_od).fillna(0).T.unstack(0).reset_index()
        
        # Extract the last 4 characters of the filename as 'ent'
        df1['ent'] = i
        
        # Rename columns and round 'flow' values to the nearest integer
        df1.columns = ['origin', 'destination', 'flow', 'ent']
        df1['flow'] = np.round(df1['flow'], 0)
        
        # Set the index to ['origin', 'destination', 'ent']
        df1 = df1.set_index(['origin', 'destination', 'ent'])
        
        # Append the current DataFrame to the list
        df_list.append(df1)

    # Concatenate all DataFrames in the list
    df_final = pd.concat(df_list, axis=0)
    
    # Reset the index and convert specific columns to 'int64' type
    df_final = df_final.reset_index()
    df_final[['origin', 'destination', 'ent']] = df_final[['origin', 'destination', 'ent']].astype('int64')
    
    return df_final

def get_assignment_matrix(assignment, Link_index, OD_index, df_final):
    """
    Create an assignment matrix with optional filtering and aggregation.

    Parameters:
        assignment (str): Path to the assignment data file (CSV).
        Link_index (list): List of Link indices to filter the assignment data.
        OD_index (list): List of OD indices to define the output shape of the matrix.
        df_final (DataFrame): DataFrame containing additional data for merging.

    Returns:
        csr_matrix: Compressed Sparse Row (CSR) matrix representing the assignment data.
    """
    # Read the assignment data from a CSV file
    assign_mat = pd.read_csv(assignment, sep=' ', header=None)
    # Define column headers for the assignment data
    headers_ = ['origin', 'destination', 'vehTypeId', 'ObjectId(section/turn)',
                'entranceIntervalIndex', 'currentIntervalIndex', 'Proportion',
                'count', 'TravelTime']
    assign_mat.columns = headers_
    # Calculate 'total_count' as the product of 'count' and 'Proportion'
    assign_mat['total_count'] = assign_mat['count'] * assign_mat['Proportion']
    # Filter assignment data based on Link indices
    assign_mat = assign_mat[assign_mat['ObjectId(section/turn)'].isin(Link_index)]
    # Merge assignment data with additional data from 'df_final'
    assign_mat = assign_mat.merge(df_final, left_on=['origin', 'destination', 'entranceIntervalIndex'],
                                  right_on=['origin', 'destination', 'ent'], how='left').sort_values(by='flow', ascending=False)
    # Calculate 'prop' as 'total_count' divided by 'flow'
    assign_mat['prop'] = assign_mat['total_count'] / assign_mat['flow']
    # Filter assignment data for specific interval conditions
    assign_mat_temp = assign_mat[(assign_mat['currentIntervalIndex'] == 1) & (assign_mat['entranceIntervalIndex'] == 1)]
    # Create a pivot table to reshape the data
    asign_ij = pd.pivot_table(assign_mat_temp, index=['ObjectId(section/turn)'],
                              columns=['origin', 'destination'], values='prop', aggfunc=np.mean).astype('float32')
    # Reindex and fill missing values with 0
    asign_ij = asign_ij.reindex(columns=OD_index, index=Link_index, fill_value=0).replace(np.nan, 0).replace(np.inf, 0)
    # Convert the DataFrame to a Cupy array
    my_array = cp.asarray(np.array(asign_ij * 0))
    # Modify a specific element in the Cupy array
    my_array[1, 1] = 0.000001
    # Create a CSR matrix from the Cupy array
    asign_ij_zero = csr_matrix(my_array).astype(cp.float32)
    # Identify unique interval combinations
    available_combos = assign_mat[['currentIntervalIndex', 'entranceIntervalIndex']].drop_duplicates(keep='first').values.tolist()
    # Initialize a list to store assignments
    assign_all = []
    # Measure time for performance evaluation
    bb = time.time()
    aa = time.time()
    # Iterate through intervals
    for i in range(0, 96):
        assign_temp = []
        print(i, bb - aa)
        aa = time.time()
        # Iterate through combinations
        for j in range(0, 96):
            if [i, j] in available_combos:
                assign_mat_temp = assign_mat[(assign_mat['currentIntervalIndex'] == i) & (assign_mat['entranceIntervalIndex'] == j)]
                asign_ij = pd.pivot_table(assign_mat_temp, index=['ObjectId(section/turn)'],
                                          columns=['origin', 'destination'], values='prop', aggfunc=np.mean, fill_value=0)
                
                asign_ij = asign_ij.reindex(columns=OD_index, index=Link_index, fill_value=0).replace(np.nan, 0).replace(np.inf, 0)
                asign_ij[asign_ij>=1] = 1
                asign_ij = csr_matrix(cp.asarray(asign_ij.values).astype(cp.float32))
                #print(asign_ij.sum().sum())
                assign_temp.append(asign_ij)
            else: 
                assign_temp.append(asign_ij_zero)
        bb = time.time()
        assign_all.append(assign_temp) 
    # Stack the assignments vertically to create a CSR matrix
    tempo = []
    for sublist in assign_all:
        a = hstack(sublist)
        tempo.append(a.tocsr())
    assignment_matrix_cp = vstack(tempo)
    return assignment_matrix_cp

def get_traffic_flow_vector_SQLITE(SQlite_file, Link_index):
    """
    Retrieve traffic flow data from an SQLite database table.

    Parameters:
        SQlite_file (str): Path to the SQLite database file.
        Link_index (list): List of Link indices to filter the traffic flow data.

    Returns:
        DataFrame: Traffic flow data for selected links.
    """
    
    with sqlite3.connect(SQlite_file) as conn:
    # Retrieve data from the 'MESECT' table in the SQLite database
        df = pd.read_sql_query("SELECT * FROM MESECT", conn)
    
    conn.close()
    # List of sections based on 'Link_index'
    list_of_sections = Link_index.to_list()
    
    # Filter the data for 'sid' equal to 1
    df = df[df['sid'] == 1]
    
    # Drop duplicate rows
    df = df.drop_duplicates(keep='first')
    
    # Filter the data for 'oid' values present in 'list_of_sections'
    df = df[df['oid'].isin(list_of_sections)]
    
    # Drop duplicate rows again
    df = df.drop_duplicates(keep='first')
    
    # Sort the DataFrame
    df.sort_values(by=['ent', 'oid'], ascending=True, inplace=True)
    
    return df

def stats(Comp, a = 'true', b = 'flow', list_special_ent =None, list_special_ent_oids = None, GEH_thresh = 5):
    if list_special_ent!= None:
        Comp = Comp[Comp['ent'].isin(list_special_ent)]
    if list_special_ent_oids!= None:
        Comp = Comp[Comp['oid'].isin(list_special_ent)]
    Comp['diff'] = Comp[a] - Comp[b]
    Comp['diff2'] = Comp['diff'] * Comp['diff']
    
    Comparison_flow_df_copy = Comp.copy(deep = True)
    Comparison_flow_df_copy = Comparison_flow_df_copy[Comparison_flow_df_copy['true'] >= 1]
    Comparison_flow_df_copy['diff'] = abs(Comparison_flow_df_copy['diff'])
    Comparison_flow_df_copy['diff_rel']= Comparison_flow_df_copy['diff']/Comparison_flow_df_copy['true']
    
    rmse = np.sqrt((Comp['diff2'].sum().sum() / (Comp.shape[0]))).round(2)
    mape = (Comparison_flow_df_copy['diff_rel'].sum().sum() * 100 / Comparison_flow_df_copy.shape[0]).round(2)
    pivot_flow = pd.pivot_table(Comp, index = 'oid', columns = 'ent', values = a)
    pivot_value = pd.pivot_table(Comp, index = 'oid', columns = 'ent', values = b)
    rolling_window = 4
    # Calculate rolling sum for every 4 columns using the rolling function
    rolling_sums_df_flow = pivot_flow.rolling(window=rolling_window, axis=1).sum()
    rolling_sums_df_flow.dropna(axis=1, inplace=True)
    rolling_sums_df_value = pivot_value.rolling(window=rolling_window, axis=1).sum()
    rolling_sums_df_value.dropna(axis=1, inplace=True)
    GEH_df = np.sqrt((2 * (rolling_sums_df_value - rolling_sums_df_flow) * (rolling_sums_df_value - rolling_sums_df_flow)) / (rolling_sums_df_value + rolling_sums_df_flow))
    GEH5 = ((GEH_df[GEH_df <= GEH_thresh]/GEH_df[GEH_df <= GEH_thresh]).replace(np.nan, 0).sum().sum() * 100/(GEH_df.shape[0] * GEH_df.shape[1])).round(2)
    return (mape, rmse, GEH5)
```

```python
def Update_G_unity_contrain_NS(G, G_prior, Wd, rho, tau, P, Weight_RDS_np, assignment_matrix_np, unity = 10000, G_prior_wt = 1):
    Ones_OD = csr_matrix(np.ones((assignment_matrix_np.shape[1], 1)))
    Ones_O = csr_matrix(np.ones((P.shape[0], 1)))
    aa = time.time()
    #print(type(Wd), type(assignment_matrix_np), type((Weight_RDS_np.multiply(RDS1_np)).dot(P.T)))
    temp1 = matrix_multiply_csr_chunked(Wd, assignment_matrix_np.T, (Weight_RDS_np.multiply(RDS1_np)).dot(P.T), chunk_size=10000)
    temp2 = Wd.multiply(rho.T.dot(temp1).dot(tau.T))
    temp3 = Weight_RDS_np.multiply(assignment_matrix_np.dot(Wd.multiply(rho.dot(Wd.multiply(G)).dot(tau)).dot(P))).dot(P.T)
    temp4 = matrix_multiply_csr_chunked(Wd, assignment_matrix_np.T, temp3, chunk_size=10000)
    temp5 = Wd.multiply((rho.T.dot(temp4).dot(tau.T)))
    temp1_constrain = Wd
    temp2_constrain = Wd.multiply(rho.T.dot(temp1_constrain).dot(tau.T))
    temp3_constrain1 = csr_matrix((Wd.multiply(rho.dot(Wd.multiply(G)).dot(tau))).sum(axis = 0))
    result_matrix = Wd.copy()
    temp4_contrain = temp3_constrain1.T.multiply(result_matrix.T).T
    temp5_contrain = Wd.multiply((rho.T.dot(temp4_contrain).dot(tau.T)))
    temp5 = temp5 + unity * temp5_contrain + G_prior_wt * G
    print(type(temp5))
    temp5.data **= -1
    second_term_G = (temp2 +  unity * temp2_constrain + G_prior_wt * G_prior).multiply(temp5)
    second_term_G = second_term_G.astype('float32')
    second_term_G = cs.csr_matrix(second_term_G)
    second_term_G.data[cp.isnan(second_term_G.data)] = 0
    second_term_G.data[second_term_G.data > 1] = (1 + 1 * (np.power(second_term_G.data[second_term_G.data > 1],1)) - 1)
    second_term_G.data[second_term_G.data < 1] = (1 + 1 * (np.power(second_term_G.data[second_term_G.data < 1],1)) - 1)
    second_term_G.data[second_term_G.data < 0] = 0
    second_term_G.data[cp.isnan(second_term_G.data)] = 0
    #second_term_G.data[second_term_G.data > 1.01] = 1.01
    G = cs.csr_matrix(G).multiply(second_term_G)
    '''G_sum = cs.csr_matrix(cp.array(G.sum(axis= 0)))
    G_sum.data = 1/G_sum.data 
    G = (cs.csr_matrix(G).multiply(G_sum))
    print(time.time() - aa)'''
    return G.get()


def update_P_NS(P, P_prior, G, assignment_matrix_np, Wd, rho, tau, Weight_RDS_np, RDS1_np, Wt_prior = 1):
    aa = time.time()
    V1 = assignment_matrix_np.dot(Wd.multiply(rho.dot(Wd.multiply(G)).dot(tau)))
    P_transpose_neum = (Weight_RDS_np.multiply(RDS1_np)).T.dot(V1) + Wt_prior * P_prior.T
    P_transpose_denom = (Weight_RDS_np.multiply(V1.dot(P))).T.dot(V1) + Wt_prior * P.T
    second_term_P = P_transpose_neum / P_transpose_denom
    second_term_P[np.isnan(second_term_P)] = 0
    second_term_P = np.where(second_term_P > 1, (1 + 1 * (np.power(second_term_P,1) - 1)), second_term_P)
    second_term_P = np.where(second_term_P < 1, (1 + 1 * (np.power(second_term_P,1) - 1)), second_term_P)
    second_term_P = np.where(second_term_P == np.nan, 0, second_term_P)
    second_term_P = np.where(second_term_P == np.inf, 0, second_term_P)
    #second_term_P = np.where(second_term_P > 1.03, 1.03, second_term_P)
    second_term_P = csr_matrix(second_term_P)
    second_term_P.data[second_term_P.data > 1.02] = 1.02
    second_term_P.sum()
    print(P.sum(), 'Sum initial')
    P = P.multiply(second_term_P.T).astype('float32')
    print(P.sum(), 'Sum later')
    #print(time.time() - aa)
    return P
```

```python

def Upper_Level_OD_Adjustmen_Non_structural(OD_cp, Weight_RDS_cp, RDS1_cp, assignment_matrix_cp, OD_cp_prior, prior_wt = 10):
    
    OD_cp_transposed = OD_cp.T # Transpose OD_cp and OD_prior_cp
    shape = OD_cp_transposed.shape # Get the shape of OD_cp_transposed
    for k in range(0,4): # Iterate through k values (0 to 4)
        numerator_c = (Weight_RDS_cp.multiply(RDS1_cp)).T.dot(assignment_matrix_cp) + prior_wt* OD_cp_prior.T # Calculate numerator_c
        denominator_c = ((Weight_RDS_cp.T.multiply(OD_cp_transposed.dot
                                                  (assignment_matrix_cp.T))).dot(assignment_matrix_cp) 
                         + prior_wt * OD_cp_transposed) # Calculate denominator_c
        second_term_c = numerator_c / denominator_c # Calculate second_term_c
        second_term_c[second_term_c > 1.03] = 1.03
        second_term_c[np.isnan(second_term_c)] = 0
        OD_cp_transposed = OD_cp_transposed.multiply(second_term_c) # Multiply OD_cp_transposed by second_term_c
        print(OD_cp_transposed.sum())
    OD_cp_transposed = OD_cp_transposed.tocsr() # Convert OD_cp_transposed to a NumPy array
    OD_cp_transposed[OD_cp_transposed >= 600] = 600 # Apply a threshold to OD_cp_transposed
    OD_cp_transposed1 = cs.csr_matrix(OD_cp_transposed) # Convert OD_cp_transposed to a CSR matrix
    OD_cp_1 = OD_cp_transposed1.T # Transpose OD_cp_transposed1 to get OD_cp_1
    sum_OD_previous = OD_cp.sum().sum() # Calculate and print some results
    sum_OD_1 = OD_cp_1.sum().sum()
    max_OD_1 = OD_cp_1.max()
    print('sum OD:', sum_OD_previous, sum_OD_1, max_OD_1)
    return OD_cp_1

def Grid_lock_modification_Non_structural(OD_df, Comparison_flow_df, assignment,  assignment_matrix_cp,
                           assignment_matrix_cp_no_grid, correction_rate = .25, 
                           threshold_flow_drop = 500):
    Comparison_flow_df['Change'] = Comparison_flow_df['true'] / Comparison_flow_df['flow']
    Comparison_flow_df_grid_blocks = Comparison_flow_df[(Comparison_flow_df['true'] >= threshold_flow_drop) & (Comparison_flow_df['Change'] == np.inf)]
    print(Comparison_flow_df_grid_blocks)
    if len(Comparison_flow_df_grid_blocks)>0:
        print('Grid_block exists')
        corrections_interval = Comparison_flow_df_grid_blocks['ent'].unique().min() - 8
        unique_oid = Comparison_flow_df_grid_blocks['oid'].unique()
    
        assign_mat = pd.read_csv(assignment, sep=' ', header=None)
        headers_ = ['origin', 'destination', 'vehTypeId', 'ObjectId(section/turn)',
                    'entranceIntervalIndex', 'currentIntervalIndex', 'Proportion',
                    'count', 'TravelTime']
        assign_mat.columns = headers_
        assign_mat_grid = assign_mat[assign_mat['ObjectId(section/turn)'].isin(unique_oid)]
        assign_mat_grid = assign_mat_grid[assign_mat_grid['entranceIntervalIndex'] > corrections_interval]
        assign_mat_grid = assign_mat_grid.drop_duplicates(subset=['origin', 'destination'], keep='first')
    
        condition = (OD_df['ent'] > corrections_interval) & (OD_df['origin'].isin(assign_mat_grid['origin'])) & (OD_df['destination'].isin(assign_mat_grid['destination']))
        OD_df.loc[condition, 0] *= correction_rate   # Reduce the values in column 0 by 25%
        OD_cp1 = cs.csr_matrix(cp.asarray(OD_df[0].values.reshape([OD_df[0].shape[0], 1])))
        replication_id = '16689389'
        Write_Updated_OD_into_the_simulation_folder(OD_cp_1, OD_vector, txt_files)
        #Simulation(console, ang_file, replication_id, assignment,object_file)
        assign = assignment_matrix_cp_no_grid.copy() #get_assignment_matrix_multiprocessing(assignment, Link_index, OD_index, OD_vector)
    if len(Comparison_flow_df_grid_blocks)== 0:
        assign = assignment_matrix_cp
        OD_df = OD_df.copy(deep = True)
    else:
        OD_df = OD_df
    return (OD_df, assign)
```
